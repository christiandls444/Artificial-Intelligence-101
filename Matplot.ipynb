```python
## tabular
import pandas as pd

## mathematical operations
import numpy as np

## visualization
import matplotlib.pyplot as plt
```

```python
var_sddf = pd.read_excel('assignment-data.xlsx', 'sales-data')

var_sddf.info()
```

```python
## display first 5 in head

var_sddf.head()
```

```python
## display first 5 in tail
 
var_sddf.tail()
```

```python
## setting parameter

var_sddf.pivot_table(

values = ['Price'],
                     
index = ['supervisor', 'company-name'],
                     
columns = 'Product-detail',
            
aggfunc = [np.sum],
                     
fill_value = 0)
```

# Answer in Question 2

```python
cbg = '#161d1a'

cbs = '#75f0bd'

var_cntq = (var_sddf.groupby('company-name').sum().reset_index().sort_values(by = 'total-quantity', ascending = False).head())

plt.title('TOP 5 COMPANIES WITH MOST ORDERS')

plt.bar(var_cntq['company-name'], var_cntq['total-quantity'], 0.5, color = cbs)

plt.xlabel('COMPANIES')

plt.ylabel('PRICE')

plt.xticks(rotation = 0)

plt.show()
```

# Answer in Question 3

```python
## setting color
cbs = '#161d1a'
cbg = '#75f0bd'

var_cnsum = var_sddf.groupby('company-name').sum()
fig, (apple, hcl, infosys) = plt.subplots(1, 3, sharex = True, sharey = True, figsize = (15, 3))

## apple
apple.set_ylabel('PRICE')
apple.set_title('APPLE')
apple.set_facecolor(cbg)
apple.set_xlabel('TOTAL QUANTITY')
apple.scatter(var_cnsum.loc['Apple']['total-quantity'], var_cnsum.loc['Apple']['Price'], color = cbs)
apple.spines['top'].set_visible(False)
apple.spines['right'].set_visible(False)

## hcl
hcl.set_title('HCL')
hcl.set_facecolor(cbg)
hcl.set_xlabel('TOTAL QUANTITY')
hcl.scatter(var_cnsum.loc['HCL']['total-quantity'], var_cnsum.loc['HCL']['Price'], color = cbs)
hcl.spines['top'].set_visible(False)
hcl.spines['right'].set_visible(False)

## infosys
infosys.set_title('INFOSYS')
infosys.set_facecolor(cbg)
infosys.set_xlabel('TOTAL QUANTITY')
infosys.scatter(var_sddf.groupby('company-name').get_group('Infosys')['total-quantity'], var_sddf.groupby('company-name').get_group('Infosys')['Price'], color = cbs)
infosys.spines['top'].set_visible(False)
infosys.spines['right'].set_visible(False)
```
