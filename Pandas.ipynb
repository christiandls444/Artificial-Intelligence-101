# Answer in Question 2 Task 2

```python
def new_created_at(created_at):
    
    created_at_list = []
    
    for date, index in zip(created_at, range(len(created_at))):
        
        if index > 2: 
        
            created_at_list.append((datetime.strptime(date, '%Y-%m-%d %H:%M:%S') - timedelta(days=1)).strftime('%A/%B/16'))
        
        else: 
            created_at_list.append((datetime.strptime(date, '%Y-%m-%d %H:%M:%S') - timedelta(days=2)).strftime('%A/%B/16'))
    
    return created_at_list

var_tdf['new_tweet_id'] = [str(id)[-5:] + str(id)[-6:4:-1] + str(id)[:5] for id in var_tdf['tweet_id']]

var_tdf['created_at4'] = new_created_at(var_tdf['created_at'])

var_tdf['username'] = var_tdf['username'].mask(lambda username: username != '', other='yyy')

var_tdf[['new_tweet_id','created_at4','username']]
```

# Answer in Question 3 Task 2

```python
var_ordf['NumberOfTransactions'] = 1

var_cisum = var_ordf.groupby('CustomerID').sum()

print("Hurray!"
      "\nThe most number of transactions is done by Customer ID: {}."
      "\nHe has made {} transactions."
      "\nHe will be rewarded with our gift hamper worth 1000 pesos."
 
.format(int(var_cisum.reset_index().set_index('NumberOfTransactions').loc[var_cisum.max()['NumberOfTransactions']]['CustomerID']),
        
int(var_cisum.max()['NumberOfTransactions'])))
```

# Answer in Question 3 Task 3

```python
var_ordf['TotalPurchaseAmount'] = var_ordf['Quantity'] * var_ordf['UnitPrice']

var_cisum = var_ordf.groupby('CustomerID').sum()

print("Customer ID: {} is the best customer since he has spent a lot on our business."
      "\nHis total purchase amount is {} pesos."
      "\n50% discount will be given on his next purchase."
      
.format(int(var_cisum.reset_index().set_index('TotalPurchaseAmount').loc[var_cisum.max()['TotalPurchaseAmount']]['CustomerID']),
    
var_cisum.max()['TotalPurchaseAmount']))
```

# Answer in Question 3 Task 4

```python
var_scsum = var_ordf.groupby('StockCode').sum()

var_siq = var_scsum.reset_index().set_index('Quantity')

var_scmax = var_siq.loc[var_scsum.max()['Quantity']]['StockCode']

var_scmin = (var_siq.loc[var_scsum.min()['Quantity']]['StockCode'].iloc[0])

print(f"Most bought item is StockCode {var_scmax} with {int(var_scsum.max()['Quantity'])} in purchased quantity.")

print(f"Least bought item is StockCode {var_scmin} with "f"{int(var_scsum.min()['Quantity'])} in purchased quantity overall."
      "\nThese two  items can be bundled and provide 10% discount on this bundle to increase the sale of Stockcode "f" {var_scmin}")
```
